```python
#Create a 1D NumPy array with integers from 1 to 20. Perform the following operations:
#  a. Calculate the sum, mean, median, and standard deviation of the elements in the array.
#  b. Find the indices of elements greater than 10 in the array.
```

```python
import numpy as np

# Create a 1D NumPy array with integers from 1 to 20
array_1d = np.arange(1, 21)
print("1D NumPy Array:", array_1d)
```

```python
# a. Calculate the sum, mean, median, and standard deviation
array_sum = np.sum(array_1d)
array_mean = np.mean(array_1d)
array_median = np.median(array_1d)
array_std_dev = np.std(array_1d)

print("\n--- OUTPUT OF cALCULATION  ARRAY_ID Array ---")
print(f"Sum: {array_sum}")
print(f"Mean: {array_mean}")
print(f"Median: {array_median}")
print(f"Standard Deviation: {array_std_dev}")
```

```python
# b. Find the indices of elements greater than 10
indices_greater_than_10 = np.where(array_1d > 10)

print("\nIndices of elements greater than 10:", indices_greater_than_10[0])
print("Elements greater than 10:", array_1d[indices_greater_than_10])
```

```python
#2. Create a 2D NumPy array of shape 4 X 4 with numbers ranging from 1 to 16.
# a. Print the array.
# b. Find the transpose of the array.
# c. Calculate the row-wise and column-wise sums of the array.
array_2d = np.arange(1, 17).reshape(4, 4)
print("Original 2D NumPy Array:")
print(array_2d)
```

```python
# b. Find the transpose of the array.
array_transpose = np.transpose(array_2d)
print("\nTranspose of the array:")
print(array_transpose)
```

```python
# c. Calculate the row-wise and column-wise sums of the array.
row_sums = np.sum(array_2d, axis=1)
column_sums = np.sum(array_2d, axis=0)
print("\nRow-wise sums:", row_sums)
print("Column-wise sums:", column_sums)
```

```python
#3. Create two 3 X 3 arrays filled with random integers between 1 and 20.
# a. Perform element-wise addition, subtraction, and multiplication.
 # b. Compute the dot product of the two arrays.
array_3d = np.random.randint(1, 21, size=(3, 3))
print("First 3D NumPy Array:")
print(array_3d)
```

```python
# a. Perform element-wise addition, subtraction, and multiplication.
array_3d_2 = np.random.randint(1, 21, size=(3, 3))
print("Second 3D NumPy Array:")
print(array_3d_2)
```

```python
# addition
addition_result = array_3d + array_3d_2
print("\nElement-wise Addition:")
print(addition_result)
#subtraction
subtraction_result = array_3d - array_3d_2
print("\nElement-wise Subtraction:")
print(subtraction_result)
#multiplication
multiplication_result = array_3d * array_3d_2
print("\nElement-wise Multiplication:")
print(multiplication_result)
```

```python
dot_product_result = np.dot(array_3d, array_3d_2)
print("\nDot Product of the two arrays:")
print(dot_product_result)
```

```python
#4. Reshape a 1D array of size 12 into a 3 X 4 2D array and slice the first two rows and last two columns.
array_2d = np.arange(1, 13).reshape(3,4)
print("2D NumPy Array:")
print(array_2d)
```

```python
# Slice the first two rows and last two columns
sliced_array = array_2d[0:2, 2:4]

print("\nSliced Array (first two rows, last two columns):")
print(sliced_array)
```

```python
slice_array = array_2d[0:-2, 2:]
print(slice_array)
```
